```python
print(df.isnull().any())
fill_zero = df.fillna(value = 0)
print(df.loc[:,['Car','BuildingArea','YearBuilt']].mean())
print(fill_zero.loc[:,['Car','BuildingArea','YearBuilt']].mean())
### I think it should not be used because the mean won't be accurate (distorted)
```

```python
impute = ['Car','BuildingArea','YearBuilt']
print(df[impute].mean())
df[impute] = df[impute].fillna(df[impute].mean())
df = df.dropna()
print(df[impute].mean())
print(df.info())
### This method should be used because the mean still be the same ( adding the mean , adding the qunatity). Also, it will not remove as much data.
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='mixed')
df['Date'] = df['Date'].dt.strftime('%Y/%m/%d')
print(df['Date'].head())
```

```python
import matplotlib.pyplot as plt

plt.pie(labels = df['Method'].unique(),x = df['Method'].value_counts(),autopct='%1.1f%%')
plt.title('Pie chart (Method)')
plt.legend()
plt.show()
```

```python
plt.bar(x=df['Rooms'].unique(),height=df['Rooms'].value_counts(),label=df['Rooms'].unique())
plt.title('Bar chart (Rooms)')
plt.xlabel('Differenct room type')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.bar(x=df['CouncilArea'].unique(),height=df['CouncilArea'].value_counts())
plt.title('Bar chart (Council Area)')
plt.xlabel('Differenct area')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
