7. [20 points] Make the following changes to the dataset.
- 7.1) [2 points] Remove an unneeded column "Unnamed: 0" *(Note: this column must no longer appear when displaying the dataframe again later)*
- 7.2) [4 points] Create a new column and name it "major_expense". This column contains values that are the sum of housing, food, and transportation.
- 7.3) [4 points] Create a new column and name it "major_expense_ratio" which is based on the following formula:

```
major_expense_ratio = (major_expense * 100) / monthly_income
```



*Note: Feel free to add more cells as needed.*

````python
# Write your code here
# 7.1) Remove an unneeded column "Unnamed: 0" (Note: this column must no longer appear when displaying the dataframe again later)
df = df.drop('Unnamed: 0',axis=1)
df
````

````python
# Write your code here
# 7.2) Create a new column and name it "major_expense". This column contains values that are the sum of housing, food, and transportation.
df['major_expense'] = df['housing'] + df['food'] + df['transportation']
df.head()
````

````python
# Write your code here
# 7.3) Create a new column and name it "major_expense_ratio"
# major_expense_ratio = (major_expense * 100) / monthly_income
df['major_expense_ratio'] = (df['major_expense']*100) / df['monthly_income']
df.head()
````

8. [8 points] Create a *pie chart* to demonstrate unique values of the attribute ```major```. In your visualization, also display chart title, percentage of distribution, and a legend.

````python
# Write your code here
import matplotlib.pyplot as plt
countt = df['major'].value_counts()
plt.pie(labels=countt.index, x= countt,autopct='%1.1f%%')
plt.title('Major proportion')
plt.legend()
plt.show()
````

9. [18 points] Create *three bar charts* to demonstrate the distribution of the following categorical features:
- Distribution of gender
- Distribution of year in school
- Distribution of preferred payment method

In your visualization, also display the chart title, a legend on both axes, and data labels. Make sure that data labels do not overlap one another.

````python
# write your code here
plt.bar(x=df['gender'].value_counts().index,height=df['gender'].value_counts(),label='count')
plt.xlabel('Different genders')
plt.ylabel('Distribution')
plt.legend()
plt.title('Gender Distribution')
plt.show()
plt.bar(x=df['year_in_school'].value_counts().index,height=df['year_in_school'].value_counts(),label='count')
plt.xlabel('Different years')
plt.ylabel('Distribution')
plt.legend()
plt.title('Years Distribution')
plt.show()
plt.bar(x=df['preferred_payment_method'].value_counts().index,height=df['preferred_payment_method'].value_counts(),label='count')
plt.xlabel('Different methods')
plt.ylabel('Distribution')
plt.legend()
plt.title('Payment Distribution')
plt.show()
````
