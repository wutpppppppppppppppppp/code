[BONUS 20 pts] Determine the percentage of customers who: 
- B1) returned the product once
- B2) returned the product at least once
- B3) never returned the product
- Finally, Plot a comparison of B2 and B3

*Note: please create additional cells to answer the above points*

```python
# Write your code here B1
customer_returned_once = superstore_order.merge(superstore_return[superstore_return["Returned"]=="Yes"],how="inner",on="Order ID")["Customer ID"].drop_duplicates(keep=False).shape[0]
customer_all = superstore_order["Customer ID"].drop_duplicates().count()
print(customer_returned_once/customer_all*100)
```

```python
# Write your code here B2
customer_returned_at_least = superstore_order.merge(superstore_return[superstore_return["Returned"]=="Yes"],how="inner",on="Order ID")["Customer ID"].drop_duplicates().shape[0]
print(customer_returned_at_least/customer_all*100)
```

```python
# Write your code here B3
bad_customer = customer_all - customer_returned_at_least
print(bad_customer/customer_all*100)
```

```python
# Write your code here B4
import matplotlib.pyplot as plt
df_bar = pd.DataFrame({
    "B2" : customer_returned_at_least,
    "B3" : bad_customer
}, index=[0])
df_bar.plot(kind='bar')
plt.ylabel('Person')
plt.title('Comparing between B2 and B3')
plt.show()
```
