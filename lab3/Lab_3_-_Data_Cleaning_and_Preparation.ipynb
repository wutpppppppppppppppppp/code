*(Hint: The Fibonacci sequence is the series of numbers where each number is the sum of the two preceding numbers. For example, 0, 1, 1, 2, 3, 5, ...)*

```python
def fibo(n):
    a, b = 0, 1
    for i in range(0, n):
        a, b = b, a + b
    return a
```

```python
# write your code here
pd.Series(data = [fibo(n) for n in range(50)])
```

### Groupby

```python
superstore_order.groupby(['Segment','Ship Mode'])[['Sales','Quantity','Discount','Profit']].sum()
```

**[Question]** Briefly describe an information that the result above conveys?

Ans: This table is a summary of the sales in all segments. In each segment is divided into difference ship mode. As can be seen, the majority sales quantity is from consumer segment. The profit significantly come from standard class ship mode from every segment. Same day ship mode is not popular like the others.

```python
superstore_order["Profit Ratio"] = superstore_order["Profit"]/superstore_order["Sales"]
```

```python
superstore_order.groupby(["Category", "Sub-Category"]).agg(mean_profit_ratio = ("Profit Ratio", "mean"))
```

**[Question]** Briefly describe an information that the result above conveys?
<br>
Ans: The table show the mean of profit ration to see which product can make a profit. As can be seen roughly, office supplies is the most profitable category. Labels is the best one that can make a profit comparing to it's sales.

### Pivot and Melt

Pivot

```python
superstore_order.pivot_table(index="State", columns="Ship Mode", values="Order ID", aggfunc="count").fillna(0).head(10)
```

```python
pivot_table_result = superstore_order.pivot_table(index="State", columns="Ship Mode", values="Order ID", aggfunc="count").fillna(0)
print(pivot_table_result)
```

Melt

```python
melted_result = pd.melt(pivot_table_result.reset_index(), id_vars=["State"], var_name="Ship Mode", value_name="Order Count")
print(melted_result)
```
